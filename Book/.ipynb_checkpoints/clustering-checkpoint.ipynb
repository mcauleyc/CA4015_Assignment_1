# Cluster Analysis

Cluster analysis or clustering is the task of grouping a set of objects in such a way that objects in the same group are more similar to each other than to those in other groups.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
from collections import Counter
from sklearn.metrics import silhouette_score
```

## The Silhouette Method


This is a method to  find the optimal number of clusters *k*. The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation). A high value is desirable. This is the formula:

\\[s(i) = \frac{b(i) - a(i)}{max(a(i), b(i))} \\]

>**_NOTE:_** *s(i)* is defined to be equal to zero if *i* is the only point in the cluster. This is to prevent the number of clusters from increasing significantly with many single-point clusters.

*a(i)* is the measure of the similarity of the point *i* to its own cluster. 

\\[a(i) = \frac{1}{\lvert C_{i}\rvert - 1} \sum_{j{\in}C_{i}, i{\neq}j} d(i,j) \\]

Similarly, *b(i)* is the measure of dissimilarity of *i* from points in other clusters.

\\[b(i) = \min_{i{\neq}j} \frac{1}{\lvert C_{j}\rvert} \sum_{j{\in}C_{j}} d(i,j) \\]

Where *d(i,j)* is the euclidean distance between the two points.


```python
def sil_value(x):
    sil = []
    kmax = 10

    # dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
    for k in range(3, kmax+1):
      kmeans = KMeans(n_clusters = k).fit(x)
      labels = kmeans.labels_
      sil.append(silhouette_score(x, labels, metric = 'euclidean'))
    
    return sil

# https://medium.com/analytics-vidhya/how-to-determine-the-optimal-k-for-k-means-708505d204eb
```

Decks A and B are known as the "good" decks because they have a positive net outcome whereas decks C and D have a negative net outcome. There are slight differences in these outcomes in each study. There are three different payloads:

1. In Fridberg, Maia, and Worthy, deck C has a variable loss of either âˆ’25, âˆ’50, or âˆ’75. 
2. In Horstmann, Streingroever, and Wetzels, they maintain the loss of deck C as a constant of -50.
3. In Kjome, Premkumar, and Wood, the schedules of rewards and losses in such a way that the net outcome of the "bad" decks decreases by 150 every 10 cards and the net outcome of the "good" decks increases by 25 every 10 cards.


For this reason I would like to look at the relationship between Deck C and the total amount won/lost and perform a cluster analysis of it.


## Performing Clustering on Deck C


I begin by making a dataframe with just the columns 'Total' and 'C'.

```python
c = df.iloc[:,[1,4]] 
c
```

I then get the silhouette value of c and from that find the optimal *k*.

```python
c_sil = sil_value(c)
c_k = c_sil.index(max(c_sil)) + 3

plt.plot([3,4,5,6,7,8,9,10], c_sil)
plt.title('Optimal k for c')
plt.xlabel('silhouette score')
plt.ylabel('k')
plt.show()
```

I then use the KMeans algorithm from sklearn and fit it to c.

```python
kmeans = KMeans(c_k)
kmeans.fit(c)
```

I then find the identified clusters in c.

```python
identified_clusters = kmeans.fit_predict(c)
identified_clusters
```

I then plot the data, with each cluster in a different colour.


```python
c_data_with_clusters = df.copy()
c_data_with_clusters['Clusters'] = identified_clusters 
# plt.scatter(c_data_with_clusters['Total'],c_data_with_clusters['C'],c=a_data_with_clusters['Clusters'],cmap='plasma')
sns.lmplot(data=c_data_with_clusters, x='Total', y='C', hue='Clusters', 
            fit_reg=False, legend=True, palette='Accent')
```

Then, for a comparison, I plot the data using colour to differentiate payload.

```python
# plt.scatter(c_data_with_clusters['Total'],c_data_with_clusters['C'],c=df['StudyNo'],cmap='plasma')
sns.lmplot(data=c_data_with_clusters, x='Total', y='C', hue='Payload', 
                   fit_reg=False, legend=True, palette='Accent')
```

As you can see the datapoints in cluster 6 all chose c less and had the highest loss in the study. All of these data points were in payload 3. This makes sense because these people obviously chose C less because they were getting frequent losses from it and those losses were growing as they went. Most of the people who chose C the most were in payload 2 which is where the losses in C were constant.

## Clustering on "good" decks and "bad" decks

Now I would like to cluster the data based on the "good" card decks and the "bad" card decks to see if there is any insight to be gleamed from it.

```python
good = df.iloc[:,[1,7]]
good_sil = sil_value(good)
good_k = good_sil.index(max(good_sil)) + 2

identified_clusters = kmeans.fit_predict(good)

plt.plot([3,4,5,6,7,8,9,10], good_sil)
plt.title('Optimal k for good')
plt.xlabel('silhouette score')
plt.ylabel('k')
plt.show()

kmeans = KMeans(good_k)
kmeans.fit(good)
good_data_with_clusters = df.copy()
good_data_with_clusters['Clusters'] = identified_clusters 
sns.lmplot(data=good_data_with_clusters, x='Total', y='Good', hue='Clusters', 
                   fit_reg=False, legend=True, palette='Accent')


sns.lmplot(data=good_data_with_clusters, x='Total', y='Good', hue='Payload', 
                   fit_reg=False, legend=True, palette='Accent')


sns.lmplot(data=good_data_with_clusters, x='Total', y='Good', hue='StudyNo', 
                   fit_reg=False, legend=True, palette='Accent')

```

As you can see from the two graphs, their is obviously a strong correlation between picking "good" decks and winning more money. There does not seem to be any correlation between payload and "good" decks though. Nothing major stands out about this study except there were less people who chose the "bad" decks, as you can see by the fact that there are fewer yellow datapoints under 0.4 than the other studies. 

I would now like to look more closely at clusters 3 and 5 because they are the most extreme on each side. It is said that after about 50 choices most people begin to see the patterns ang go for the "good" decks more consistently. I want to see if the people in cluster 5 seemed to figure out the pattern and the people in cluster 3 did not.

```python
good_cluster = good_data_with_clusters[good_data_with_clusters['Total'] == max(good_data_with_clusters['Total'])]['Clusters'].iloc(0)[0]
bad_cluster = good_data_with_clusters[good_data_with_clusters['Total'] == min(good_data_with_clusters['Total'])]['Clusters'].iloc(0)[0]
```

```python
good_subj_3 = list(good_data_with_clusters[good_data_with_clusters['Clusters'] == bad_cluster]['Subj'])
good_subj_3

good_lst = []
for i in good_subj_3:
    n, idx = i.split('_', 1)
    if n == '100':
        good_lst.append(list(all_100[all_100.index == idx].values[0])[101::2])
#         print(list(all_100[all_100.index == idx].values[0])[-100::2])
    elif n == '150':
        good_lst.append(list(all_150[all_150.index == idx].values[0])[101::2])
#         print(list(all_150[all_150.index == idx].values[0])[-150::2])
    elif n == '95':
        good_lst.append(list(all_95[all_95.index == idx].values[0])[101::2])

good_d = {'a' : 0, 'b' : 0, 'c' : 0, 'd' : 0}
good_l, no_bad = [], []
for i in good_lst:
    c = Counter(i)
    good_d['a'] += c[1]
    good_d['b'] += c[2]
    good_d['c'] += c[3]
    good_d['d'] += c[4]
    if c[1] + c[2] != 0:
        good_l.append((c[3] + c[4]) / (c[1] + c[2]))
    else:
        no_bad.append(c)
print(good_d)
print(len([x for x in good_l if x < .5]))
print(len(good_l))
print(no_bad)
print(np.mean(good_l))
```

This cluster chose b the most and 6 out of 8 people were twice as likely to choose "bad" decks over "good" decks. This shows that the people in this cluster probably did not see the pattern and got distracted by the infrequent losses in Deck B.

The mean of the rations between good and bad is 0.369, which shows that the participants in the cluster were far less likely to choose good decks. If the mean was 1 it would mean that the participants chose fairly evenly between the "good" and "bad" decks.
